### Visualizations

```python
age_absent = df[['Age', 'Absenteeism time in hours']].sort_values('Absenteeism time in hours', ascending=False)
age_absent
```

```python
absencesperage = df.groupby('Age')['Absenteeism time in hours'].sum()
absencesperage
```

```python
fig, ax = plt.subplots()
absencesperage.plot.bar(ax=ax, stacked=True)
plt.ylabel('Hours')
plt.title('Total hours of absence per age')
plt.xticks(rotation=60)
fig.tight_layout()
```

#### From the graph above we can see that the 28 age group had the highest amount of absences in hours.

```python
monthabsent = df['Month of absence'].value_counts().head(5)
monthabsent
```

```python
fig, ax = plt.subplots()
monthabsent.plot.pie()
plt.title('Months employees were absent in ratio')
```

#### From the graph above we can see that the top 5 months for absences were February, March, May, July and October.

```python
df['Hit target']
```

```python
agetarget = df.groupby('Age')['Service time'].sum()
agetarget
```

```python
fig, ax = plt.subplots()
agetarget.plot.line()
plt.title('Total service time by age')
plt.ylabel('Hours')
fig.tight_layout()
```

#### From the graph above we can see that the most service time was completed by workers aged between 30 to 40.

```python
dayabsent = df['Day of the week'].value_counts().head()
dayabsent
```

```python
fig, ax = plt.subplots()
dayabsent.plot.bar()
plt.xlabel('Day of the week')
plt.ylabel('Total days absent')
plt.title('Days of absence')
plt.xticks(rotation=0)
fig.tight_layout()
```

#### From the bar graph above we can see that the second day of the week had the highest day of absence.
